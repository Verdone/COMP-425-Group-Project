```python
#"".join([x[1] for x in sorted([[box.xyxy.tolist()[0][0],class_names[int(box.cls[0])]] for box in results[0].boxes])])
def run_ocr(plate,model,class_names,threshold):
    results = model(plate)
    #class_name = class_names[int(results[0].boxes[0].cls[0])]
    if len(results[0].boxes[:].conf) == 0:
        return None
    conf = sum(results[0].boxes[:].conf)/len(results[0].boxes[:].conf)
    if conf>threshold:
        return "".join([x[1] for x in sorted([[box.xyxy.tolist()[0][0],class_names[int(box.cls[0])]] for box in results[0].boxes])])
    else:
        return None

```

```python
def crop_image(path,box):
    x1, y1, x2, y2 = box.xyxy[0]
    image = cv2.imread(path)
    x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
    cropped = image[y1:y2, x1:x2,:]
    return cropped
```

```python
def annotate_image_with_label(
    image,
    x1: int,
    y1: int,
    x2: int,
    y2: int,
    label: str
):

    # 1) Load the image


    if image is None:
        raise ValueError(f"Error loading image at {image_path}")

    cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)

    text_position = (x1, max(0, y1 - 10))  # ensure we don't draw text off-image
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 0.6
    text_color = (0, 0, 255)   # green text
    thickness = 2

    cv2.putText(image, label, text_position, font, font_scale, text_color, thickness)

    return image

```

```python
image_extensions = {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff"}
input_folder = r"/content/container/valid/images"
is_break=False
for filename in os.listdir(input_folder):
    # Check file extension to ensure it's an image
    ext = os.path.splitext(filename)[1].lower()
    if ext in image_extensions:
        # Construct full file path
        color_image_path = os.path.join(input_folder, filename)

        image=cv2.imread(color_image_path)
        results = model(color_image_path)
        result=results[0]
        for box in result.boxes:
            cropped=crop_image(color_image_path,box)
            plate_chars = run_ocr(cropped,model2,class_names,0.99)
            if plate_chars:
              image = annotate_image_with_label(image,int(box.xyxy[0][0]),int(box.xyxy[0][1]),int(box.xyxy[0][2]),int(box.xyxy[0][3]),plate_chars)
              is_break=True
            else:
              image = annotate_image_with_label(image,int(box.xyxy[0][0]),int(box.xyxy[0][1]),int(box.xyxy[0][2]),int(box.xyxy[0][3]),"Unknown")
        if is_break:
          cv2_imshow(image)
          break

```
